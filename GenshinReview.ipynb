```python
import pandas as pd
import string 
df = pd.read_csv('./output.csv')
df.head()
```

```python
#remove useless columns
df= df.drop(columns=['reviewId','at','appVersion'], axis=1)
df.head()
```

```python
#converting them into all lowercase
df['clean_content'] = df['content'].str.lower()
df.head()
```

```python
string.punctuation
```

```python
#function to remove punctuations
def remove_punctuations(text):
    punctuations = string.punctuation
    return text.translate(str.maketrans('', '', punctuations))
```

```python
#applying the function into the current clean_content
df['clean_content'] = df['clean_content'].apply(lambda x: remove_punctuations(x))
df.head()
```

```python
#stopwords
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
", ".join(stopwords.words('english'))
```

```python
#removing stop word
stopWords = set(stopwords.words('english'))
def remove_stopwords(text):
    return " ".join([word for word in text.split() if word not in stopWords])
```

```python
df['clean_content'] = df['clean_content'].apply(lambda x: remove_stopwords(x))
df.head()
```

```python
from collections import Counter

#counting num of most common words first
word_count = Counter()
for text in df['clean_content']:
    for word in text.split():
        word_count[word] += 1
        
word_count.most_common(10)
```

```python
#remove frequent words
freqWords = set(word for (word, wc) in word_count.most_common(3))
def remove_freq_words(text):
    return " ".join([word for word in text.split() if word not in freqWords])
```

```python
df['clean_content'] = df['clean_content'].apply(lambda x: remove_freq_words(x))
df.head()
```

```python
#do the same for rare words
rareWords = set(word for (word, wc) in word_count.most_common()[:-10:-1])
rareWords
```

```python
def removeRareWords(content):
    return " ".join([word for word in content.split() if word not in rareWords])
```

```python
df['clean_content'] = df['clean_content'].apply(lambda x: removeRareWords(x))
df.head()
```

```python
#removing special characters
import re
def removeSpecialChars(text):
    text = re.sub('[^a-zA-Z0-9]', ' ', text)
    text = re.sub('\s+', ' ', text)
    return text
```

```python
df['clean_content'] = df['clean_content'].apply(lambda x: removeSpecialChars(x))
df.head()
```

```python
#remove urls
def remove_url(text):
     return re.sub(r'https?://\S+|www\.\S+', '', text)
```

```python
df['clean_content'] = df['clean_content'].apply(lambda x: remove_url(x))
df.head(30)
```

```python
#remove html tags
def remove_html_tags(text):
    return re.sub(r'<.*?>', '', text)
```

```python
df['clean_content'] = df['clean_content'].apply(lambda x: remove_url(x))
df.head(30)
```
